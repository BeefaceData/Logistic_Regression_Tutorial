```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv('ages_sections.csv')
df
```

```python
# convert the string values to numerical values using OrdinalEncoder
age_range = ['young', 'old']
enc = OrdinalEncoder(categories=[age_range])
df['section'] = enc.fit_transform(df[['section']])
df
```

```python
plt.scatter(df['age'],df['section'])
plt.show()
```

```python
sns.countplot(x='section', data=df)
```

```python
# setting the label X and target y
X = df.iloc[:,0:1]
y = df.iloc[:,1]
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
print(X_train.shape, X_test.shape)
```

```python
# create our ML model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# set prediction model
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
